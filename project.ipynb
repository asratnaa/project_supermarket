```python
from tabulate import tabulate
class Transaction:
    def __init__(self,id):
        self.id = id
        self.items = []

    def add_item(self,nama_item,jumlah_item,harga_barang):
        cart_item = {"nama_item":nama_item, "jumlah_item" : jumlah_item , "harga_barang" : harga_barang}
        self.items.append(cart_item)
        return cart_item
        
    def update_item_name(self,nama_baru):
        for nama in self.items:
            nama["nama_item"] = nama_baru   
        return self.items    

    def update_item_qty(self,qty_baru):
        for qty in self.items:
            qty["jumlah_item"] = qty_baru   
        return self.items 

    def update_item_price(self,price_baru):
        for price in self.items:
            price["jumlah_item"] = price_baru  
        return self.items  
        

    def delete_item(self,remove_item):
        for remove in self.items:
            if remove["nama_item"] == remove_item:
                self.items.remove(remove)
                print(self.items)
                return f'{remove_item} removed from the shopping cart.'
                
            
        return f'produk tidak ada di keranjang'    

    def reset_transaction(self):
        self.items.clear()
        print(self.items)
        return "Semua item telah dihapus dari keranjang."    
    
    def check_order(self):
        for data in self.items:
            if type(data["nama_item"]) == str:
                if type(data["jumlah_item"]) == int:
                    if type(data["harga_barang"]) == int:
                        return( "data pesanan sudah sesuai")
                    else:
                        return( "terdapat kesalahan input data")
                else:
                        return("terdapat kesalahan input data")
            else:
                        return("terdapat kesalahan input data")

           
        
                
            
            




    def total_price(self):
        total = sum(x["harga_barang"] for x in self.items)
        harga_setelah_discount = 0

        if total > 500_000:
            harga_setelah_discount += total - (total*0.1)  
        elif total > 300_000:
            harga_setelah_discount += total - (total*0.08)

        elif total > 200_000:
            harga_setelah_discount += total - (total*0.05)  

        else:
            harga_setelah_discount += total       

        return harga_setelah_discount
    
    def show_cart(self):
        if not self.items:
            return "Keranjang Kosong"
        
        table_cart = [] 
        for i,item in enumerate  (self.items, start = 1) :
            table_cart.append([i, item["nama_item"], item["jumlah_item"], item["harga_barang"],self.total_price()])

        headers = ["No", "Nama Item", "Jumlah Item", "Harga/Item", "Total Harga"]
        return tabulate(table_cart,headers= headers , tablefmt= "grid")   

        
    


trnsct_123 = Transaction("123raara")
print(trnsct_123.id)

data = trnsct_123.add_item("komputer",12,12000)
# print(data)
data2 = trnsct_123.add_item("komputer",12,120000)

        
nama_salah = trnsct_123.update_item_name("hp")
# print(nama_salah)     

qty = trnsct_123.update_item_qty(10)
# print(qty)     

price = trnsct_123.update_item_price(10)
# print(price) 

# cancel = trnsct_123.delete_item("hp")
# cancel

# reset = trnsct_123.reset_transaction()
# reset

cek = trnsct_123.check_order()
cek

total = trnsct_123.total_price()
total

trnsct_123.show_cart()
```
