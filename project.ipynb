```python
from tabulate import tabulate
class Transaction:
    def __init__(self,id):
        self.id = id
        self.items = []
        self.counter = 1

    def add_item(self,nama_item,jumlah_item,harga_barang,kategori):
        if not isinstance(nama_item,str) or not isinstance (jumlah_item,int) or not isinstance(harga_barang,(int,float)):
            return "Error: tipe data tidak sesuai."
        if jumlah_item <= 0 or harga_barang <= 0:
            return "Error: Jumlah barang dan harga harus lebih dari 0."
        cart_item = {"id":self.counter,"nama_item":nama_item, "jumlah_item" : jumlah_item , "harga_barang" : harga_barang , "Kategori_barang": kategori}
        self.items.append(cart_item)
        self.counter += 1
        return cart_item
        
    def update_item_name(self,nama_lama,nama_baru):
        for nama in self.items:
            if nama["nama_item"] == nama_lama: 
                nama["nama_item"] = nama_baru  
                return f'nama item : {nama_lama} telah di ubah menjadi {nama_baru}' 
        return f'nama item tidak ditemukan'    

    def update_item_qty(self,nama_lama,qty_baru):
        for qty in self.items:
            if qty["nama_item"] == nama_lama: 
                qty["jumlah_item"] = qty_baru  
                return f'jumlah item : {nama_lama} telah di ubah menjadi {qty_baru}' 
        return f'quantity item tidak ditemukan' 

    def update_item_price(self,nama_lama,price_baru):
        for price in self.items:
            if price["nama_item"] == nama_lama:  
                price["harga_barang"] = price_baru  
                return f'price item : {nama_lama} telah di ubah menjadi {price_baru}' 
        return f'price item tidak ditemukan'   
        return self.items  
        

    def delete_item(self,remove_item):
        for remove in self.items:
            if remove["nama_item"] == remove_item:
                self.items.remove(remove)
                print(self.items)
                return f'{remove_item} removed from the shopping cart.'        
            
        return f'produk tidak ada di keranjang'    

    def reset_transaction(self):
        self.items.clear()
        self.counter = 1
        return "Semua item telah dihapus dari keranjang."    
    
    def check_order(self):
        for item in self.items:
            if not isinstance(item["nama_item"], str) or not isinstance(item["jumlah_item"], int) or not isinstance(item["harga_barang"], (int,float)):
                return "Terdapat kesalahan input data."
        return "Data pesanan sudah sesuai."


    def total_price(self):
        total = sum(x["harga_barang"] * x["jumlah_item"] for x in self.items)
        harga_setelah_discount = total

        if total > 500_000:
            harga_setelah_discount -= total * 0.1 
        elif total > 300_000:
            harga_setelah_discount -= total *0.08

        elif total > 200_000:
            harga_setelah_discount -= total *0.05

              

        return harga_setelah_discount
    
    def show_cart(self):
        if not self.items:
            return "Keranjang Kosong"
        
        table_cart = [] 
        for i,item in enumerate  (self.items, start = 1) :
            total_per_item = item["jumlah_item"] * item["harga_barang"]
            table_cart.append([i,item["id"], item["nama_item"], item["jumlah_item"], item["harga_barang"],total_per_item])

        headers = ["No","ProductID", "Nama Item", "Jumlah Item", "Harga/Item", "Total Harga"]
        cart_table = tabulate(table_cart,headers= headers , tablefmt= "grid")   

        total_harga_akhir = self.total_price()  # Total harga setelah diskon
        return f"{cart_table}\n\nTotal Harga Keseluruhan (setelah diskon): Rp {total_harga_akhir}"
    

```

```python

# Creating a ID transaction ID
trnsct_123 = Transaction("123raara")

# Display the initial items
print(f"Initial items: {trnsct_123.items}")


# Adding items to the cart
item1 = trnsct_123.add_item(nama_item = "komputer",
                            jumlah_item= 2, harga_barang= 12_000_000,
                            kategori="elektronik")
item2 = trnsct_123.add_item(nama_item = "airfryer",
                            jumlah_item= 1, harga_barang= 420_000,
                            kategori="elektronik")

item3 = trnsct_123.add_item(nama_item = "buah naga",
                            jumlah_item= 3, harga_barang= 50_000,
                            kategori="buah")


#update items name         
nama_salah = trnsct_123.update_item_name(nama_lama="komputer",nama_baru="hp")
print(nama_salah)     

#update qty
qty = trnsct_123.update_item_qty(nama_lama="komputer",qty_baru=5)
print(qty)     

#update item price
price = trnsct_123.update_item_price(nama_lama="buah naga",price_baru=45_000)
print(price) 

#cancel item in cart
cancel = trnsct_123.delete_item("airfryer")
print(cancel)

#hapus semua barang di cart
reset = trnsct_123.reset_transaction()
print(reset)


item4 = trnsct_123.add_item(nama_item = "vaseline",
                            jumlah_item= 1, harga_barang= 100_000,
                            kategori="kecantikan")
item5 = trnsct_123.add_item(nama_item = "blackmores",
                            jumlah_item= 5, harga_barang= 200_000,
                            kategori="vitamin")

#cek order
cek = trnsct_123.check_order()
print(cek)

#cek total price
total = trnsct_123.total_price()
print(total)

#cek isi cart
print(trnsct_123.show_cart())
```
